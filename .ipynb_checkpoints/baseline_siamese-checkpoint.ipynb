```python
from tensorflow.keras.layers import Input, Conv2D, MaxPooling2D, Flatten, Dense, GlobalMaxPooling2D, subtract, BatchNormalization
from tensorflow.keras.models import Model
from tensorflow.keras.regularizers import l1, l2
from tensorflow.keras.optimizers import Adam
import tensorflow.keras.backend as K

# baseline model

input_shape = (256, 256, 3)

# conv base model
img_input = Input(input_shape)
cnn1 = Conv2D(64,(3,3),activation='relu',input_shape=input_shape, kernel_regularizer=l2(2e-4))(img_input)
mp1 = MaxPooling2D()(cnn1)
mp1 = BatchNormalization()(mp1)
cnn2 = Conv2D(128,(3,3),activation='relu', kernel_regularizer=l2(2e-4))(mp1)
mp2 = MaxPooling2D()(cnn2)
mp2 = BatchNormalization()(mp2)
cnn3 = Conv2D(128,(3,3),activation='relu', kernel_regularizer=l2(2e-4))(mp2)
mp3 = MaxPooling2D()(cnn3)
mp3 = BatchNormalization()(mp3)
cnn4 = Conv2D(256,(3,3),activation='relu', kernel_regularizer=l2(2e-4))(mp3)
mp4 = MaxPooling2D()(cnn4)
mp4 = BatchNormalization()(mp4)
flat = Flatten()(mp4)
embed = Dense(64, activation="sigmoid")(flat)

conv_base = Model(img_input, embed)

conv_base.summary()
```

```python

left_input = Input(input_shape)
right_input = Input(input_shape)

# encode each of the two inputs into a vector with the base conv model
encoded_l = conv_base(left_input)
encoded_r = conv_base(right_input)
# merge two encoded inputs with the l1 distance between them

diff = subtract([encoded_l,encoded_r])
l1_diff = K.abs(diff)
prediction = Dense(1, activation='sigmoid')(l1_diff)

siamese_net = Model([left_input,right_input],prediction)

optimizer = Adam(0.001)

siamese_net.compile(loss="binary_crossentropy",optimizer=optimizer)

siamese_net.summary()
```

```python
def pair_generation(X_p, X_n): # a small batch
    # given positive and negative samples, generate pairs
    x1 = []
    x2 = []
    y = []
    
    for i in range(len(X_p)):
        for j in range(i+1, len(X_p)):
            x1.append(X_p[i]/255.)
            x2.append(X_p[j]/255.)
            y.append(1.0)
            
    for i in range(len(X_p)):
        for j in range(len(X_n)):
            if random.randint(0,1) == 0:
                x1.append(X_p[i]/255.)
                x2.append(X_n[j]/255.)
                y.append(0.0)
            else:
                x1.append(X_n[j]/255.)
                x2.append(X_p[i]/255.)
                y.append(0.0)
            
    for i in range(len(X_n)):
        for j in range(i+1, len(X_n)):
            x1.append(X_n[i]/255.)
            x2.append(X_n[j]/255.)
            y.append(1.0)
            
    x1 = np.array(x1)
    x2 = np.array(x2)
    y = np.array(y)

            
    return x1, x2, y
```

```python
import math
from tensorflow.keras.callbacks import LearningRateScheduler
def step_decay(epoch):
    initial_lrate = 0.01
    drop = 0.5
    epochs_drop = 10.0
    lrate = initial_lrate * math.pow(drop,  
           math.floor((1+epoch)/epochs_drop))
    if lrate <= 0.000001:
        return 0.000001
    return lrate

lrate = LearningRateScheduler(step_decay)
```

```python
plt.plot([step_decay(a) for a in range(100)])
plt.show()
```

```python
def k_shot_learning(model, k = 1, epochs = 5, batch_size = 1):
    batch_size = k
    X_p = pos[:k,:,:,:]
    X_n = neg[:k,:,:,:]
    
    x1, x2, y = pair_generation(X_p, X_n)
    
    hist = model.fit([x1, x2], y, epochs = epochs, batch_size = batch_size, callbacks=[lrate])
    
    return model, hist    
```

```python
from tqdm import tqdm_notebook
def k_acc(model, anchor_p = None, anchor_n = None, k = 1):
    p_acc = 0.0
    p_cor = 0
    p_tot = 0
    # by default, anchor = 0th image
    for i in tqdm_notebook(range(k,pos.shape[0])):
        p = model.predict([pos[0].astype('float32').reshape(1,256,256,3)/255., pos[i].astype('float32').reshape(1,256,256,3)/255.])[0,0]
        if p > 0.5:
            p_cor += 1
        p_tot += 1
        

    p_acc = p_cor/p_tot
    
    print(p_acc)
        
    n_acc = 0.0
    n_cor = 0
    n_tot = 0
    # by default, anchor = 0th image
    for i in tqdm_notebook(range(k,neg.shape[0])):
        p = model.predict([neg[0].astype('float32').reshape(1,256,256,3)/255., neg[i].astype('float32').reshape(1,256,256,3)/255.])[0,0]
        if p > 0.5:
            n_cor += 1
        n_tot += 1
        
    n_acc = n_cor/n_tot
    
    print(n_acc)
        
    final_acc = 0.
        
    final_acc = (p_cor+n_cor)/(p_tot+n_tot)
        
    return final_acc
```
